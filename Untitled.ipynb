```python
import re
import pandas as pd
```

```python
with open("results.txt") as f:
    firstver = f.readlines()
```

```python
results = {"sidelen" : [int(re.findall(r'[0-9]+', firstver[i])[0]) for i in range(0, len(firstver), 3)],
"thruput" : [float(re.findall(r'[0-9].+', firstver[i])[0]) for i in range(1, len(firstver), 3)],
"period" : [float(re.findall(r'[0-9].+', firstver[i])[0]) for i in range(2, len(firstver), 3)]}
results = pd.DataFrame(results)
results.sort_values(by='sidelen', inplace=True)
results
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')

plt.plot(results.sidelen, results.period)
plt.show()
```

```python
res = results[results.sidelen > 200]
plt.plot(res.sidelen, res.thruput)
plt.show()
```
